```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housingData = pd.read_csv('D:/My Files/EDA/HousingData.csv') 
columns = housingData.columns
rows = housingData.sample
print("The size of the enitre dataset including rows and columns: ", housingData.size)
print("The rows and columns of the dataset: ", housingData.shape)
sampleRecord = housingData.head()
print(sampleRecord.info)
print(sampleRecord.describe)
```

```python
for i in columns:
    if(housingData[columns].isnull == True):
        housingData[columns].fillna(housingData[columns].mean, inplace = True)
```

```python
plt.figure(figsize = (10,5))
x = housingData['AGE']
y = housingData['MEDV']
plt.scatter(x, y)
plt.title("Scatter Plot of Ages of building and Price of the House")
plt.xlabel('Age of the building')
plt.ylabel('Price of Building')
```

```python
plt.figure(figsize = (10,5))
plt.hist(data = housingData, x = 'MEDV', bins = 10, color = 'red', edgecolor = 'black') 
plt.title('Histogram of Price of building')
plt.xlabel('Price of building')
plt.ylabel('Price Frame')
```

```python
plt.figure(figsize = (10,5))
plt.title('Dense Plot of Average Number of Rooms and Price of the House')
plt.xlabel('Average Number of Rooms per House')
plt.ylabel('Price of the house')

x = housingData['RM'] 
y = housingData['MEDV']

plt.hist2d(x, y, bins = 2, cmap = 'viridis')
plt.colorbar(label = 'Density')
```

```python
plt.figure(figsize = (10, 5))
plt.title('Bar graph of the relationship between Access to Highways and Price of the House')
plt.xlabel('Accessibility to Highways')
plt.ylabel('Price of the House')

x = housingData['RAD']
y = housingData['MEDV']

plt.bar(x, y)
```

```python
plt.figure(figsize = (10, 5))
plt.title('Violin Plot between Pupil-Teacher Ratio and Price of the House')
plt.xlabel('Pupil-Teacher Ratio')
plt.ylabel('Price of House')

housingData['PTRATIO-Binned'] = pd.cut(housingData['PTRATIO'], bins = 3)

sns.violinplot(data = housingData, x = 'PTRATIO-Binned', y = 'MEDV')
```

```python
plt.figure(figsize=(10, 5))
plt.title('Box Plot between Distance to Boston Employment Centre and Price of House')
plt.xlabel('Distance to Boston Employment Centre')
plt.ylabel('Price of the House')

housingData['DIS-Binned'] = pd.cut(housingData['DIS'], bins = 3)

sns.boxplot(data=housingData, x = 'DIS-Binned', y = 'MEDV')
```

```python
list = housingData['TAX']
max = list[0]
min = list[0]

for i in list:
    if(i > max):
        max = i

for i in list:
    if(i < min):
        min = i

labels = ['100 - 199', '200 - 299', '300 - 399', '400 - 499', '600 - 699', '700 - 799']
sliceOne = []
sliceTwo = []
sliceThree = []
sliceFour = []
sliceFive = []
sliceSix = []
sliceSeven = []
sliceDivisions = []

for i in housingData["TAX"]:
    if(i >= 100 and i < 200):
        sliceOne.append(i)
    
    elif(i >= 200 and i < 300):
        sliceTwo.append(i)

    elif(i >= 300 and i < 400):
        sliceThree.append(i)

    elif(i >= 400 and i < 500):
        sliceFour.append(i)

    elif(i >= 500 and i < 600):
        sliceFive.append(i)

    elif(i >= 600 and i < 700):
        sliceSix.append(i)

    elif(i >= 700 and i < 800):
        sliceSeven.append(i)

sliceDivisions.append(sum(sliceOne))
sliceDivisions.append(sum(sliceTwo))
sliceDivisions.append(sum(sliceThree))
sliceDivisions.append(sum(sliceFour))
# sliceDivisions.append(sum(sliceFive))
sliceDivisions.append(sum(sliceSix))
sliceDivisions.append(sum(sliceSeven))

sliceDivisions[0] = sliceDivisions[0] / len(sliceOne)
sliceDivisions[1] = sliceDivisions[1] / len(sliceTwo)
sliceDivisions[2] = sliceDivisions[2] / len(sliceThree)
sliceDivisions[3] = sliceDivisions[3] / len(sliceFour)
# sliceDivisions[4] = sliceDivisions[4] / len(sliceFive)
sliceDivisions[4] = sliceDivisions[4] / len(sliceSix)
sliceDivisions[5] = sliceDivisions[5] / len(sliceSeven)

# print(sliceDivisions)
# print(sliceOne)
# print(sliceTwo)
# print(sliceThree)
# print(sliceFour)
# print(sliceFive)
# print(sliceSix)
# print(sliceSeven)

plt.figure(figsize=(10, 5))
plt.title('Pie Chart of Tax Divisons')
plt.pie(sliceDivisions, labels=labels, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'})
```

```python
plt.figure(figsize=(10, 5))
plt.title('Correlation Heat Map of the Boston Housing Dataset')

housingData = housingData.select_dtypes(include=[np.number])
housingData = pd.get_dummies(housingData, drop_first=True)
correlationMatrix = housingData.corr()
sns.heatmap(correlationMatrix, annot=True, cmap='coolwarm')
```

```python
plt.figure(figsize=(10, 5))
plt.title('Line Graph of Crime Rate in Residential Area')
plt.xlabel('Proportion of Residential Area')
plt.ylabel('Crime Rate')

# housingData = housingData.select_dtypes(include=[np.number])
# housingData = pd.get_dummies(housingData, drop_first=True)
sorted_data = housingData.sort_values(by='NOX')
x = sorted_data['NOX']
y = sorted_data['ZN']
y_smoothed = y.rolling(window=5).mean()
plt.plot(x, y_smoothed, marker='o', linestyle='-', color='b')
plt.grid(True, linestyle='--', alpha=0.6)
```
